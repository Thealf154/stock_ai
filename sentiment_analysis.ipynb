```python
import pandas as pd
import os
from textblob import TextBlob
```

```python
path = os.path.join("./", "results/", "output2.csv")
df = pd.read_csv(path)
df.tail()
```

```python
def get_sentiment(row):
    sentiments_title = TextBlob(row['title']).sentiment
    sentiments_description = TextBlob(row['description']).sentiment
    results = list(sentiments_title)
    results.extend(list(sentiments_description))
    return results

df[[
    'polarity_title','subjectivity_title',
    'polarity_description','subjectivity_description'
]] = df.apply(lambda row: pd.Series(get_sentiment(row)), axis=1)
df
```

```python
path = os.path.join("./", "news_sentiments.csv")
df.to_csv(path, index=False)
```
