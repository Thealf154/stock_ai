```python
import pandas as pd
import os
```

```python
columns = 'Ticker','Date','Open','High','Low','Close','Volume','Dividends','Stock Splits'
stocks = pd.DataFrame(columns=columns)
TICKERS = [
    'AAPL', 'MSFT', 'GOOG', 'AMZN', 'NVDA', 'META',
    'TSLA', 'BABA', 'CRM', 'AMD', 'INTC', 'PYPL',
    'ATVI', 'EA', 'TTD', 'ZG', 'MTCH', 'YELP',
]
for ticker in TICKERS:
    stock_path = os.path.join("./", "results2/", ticker + '.csv')
    stock = pd.read_csv(stock_path)
    stock["Ticker"] = ticker
    stocks = pd.concat([stocks, stock], sort=False)
    
stocks
```

```python
stocks.to_csv("./stocks.csv", index=False)
```
